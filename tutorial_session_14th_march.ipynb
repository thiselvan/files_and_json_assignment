```python
"""You are working as a data analyst at a retail company. Your manager has asked you to analyze customer purchasing behavior using a transactional orders database. You will load two real-world datasets into an in-memory SQLite database using Python, and write SQL queries to answer business questions involving aggregation, filtering, and table joins.
"""
```

```python
import pandas as pd
import sqlite3
```

DataSet Links

```python
customers_url = " https://raw.githubusercontent.com/graphql-compose/graphql-compose-examples/master/examples/northwind/data/csv/customers.csv"

orders_url = "https://raw.githubusercontent.com/graphql-compose/graphql-compose-examples/master/examples/northwind/data/csv/orders.csv"
```

```python
customers_df = pd.read_csv(customers_url)
orders_df = pd.read_csv(orders_url)
```

```python
conn = sqlite3.connect(":memory:")
```

```python
from numpy._core.defchararray import index
customers_df.to_sql("customers" , conn , index = False , if_exists="replace")
orders_df.to_sql("orders" , conn, index = False , if_exists="replace")
```

```python
query1 = """
SELECT
    CustomerID,
    COUNT(OrderID)     AS order_count,
    SUM(Freight)       AS total_freight,
    AVG(Freight)       AS avg_freight
FROM orders
GROUP BY CustomerID
ORDER BY total_freight DESC;
"""
result1 = pd.read_sql_query(query1, conn)
print(result1.head(10))
```

```python
query_a = """
SELECT CustomerID, COUNT(OrderID) AS high_freight_orders
FROM orders
WHERE Freight > 50
GROUP BY CustomerID;
"""

result_a = pd.read_sql_query(query_a, conn)
```

```python
result_a
```

```python
query_b = """
SELECT CustomerID, SUM(Freight) AS total_freight
FROM orders
GROUP BY CustomerID
HAVING SUM(Freight) > 500;
"""


result_b = pd.read_sql_query(query_b, conn)
```

```python
query_inner = """
SELECT
    c.CompanyName,
    c.Country,
    COUNT(o.OrderID)  AS order_count,
    SUM(o.Freight)    AS total_freight
FROM customers c
INNER JOIN orders o ON c.CustomerID = o.CustomerID
GROUP BY c.CompanyName, c.Country
ORDER BY total_freight DESC;
"""
```

```python
result_inner = pd.read_sql_query(query_inner, conn)
```

```python
result_inner
```

```python
query_left = """
SELECT
    c.CompanyName,
    c.Country,
    COUNT(o.OrderID)  AS order_count,
    SUM(o.Freight)    AS total_freight
FROM customers c
LEFT JOIN orders o ON c.CustomerID = o.CustomerID
GROUP BY c.CompanyName, c.Country
ORDER BY total_freight DESC;
"""
```

```python
result_left  = pd.read_sql_query(query_left, conn)
```
