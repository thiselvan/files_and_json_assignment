You are a data analyst at an e-commerce company. Your manager needs insights about sales performance across different regions, product categories, and salespersons. You have been asked to analyze the sales data to identify top performers and generate summary reports.

Since you cannot access the company database directly, create a synthetic sales dataset using the code below, then complete all analysis tasks.



```python
import pandas as pd
import random

random.seed(42)

# Define data parameters
regions = ['North', 'South', 'East', 'West']
categories = ['Electronics', 'Clothing', 'Home & Garden', 'Sports', 'Books']
salespersons = ['Alice', 'Bob', 'Carol', 'David', 'Emma', 'Frank']

# Generate 200 sales transactions
data = {
    'transaction_id': range(1001, 1201),
    'region': [random.choice(regions) for _ in range(200)],
    'category': [random.choice(categories) for _ in range(200)],
    'salesperson': [random.choice(salespersons) for _ in range(200)],
    'sales_amount': [round(random.uniform(50, 5000), 2) for _ in range(200)],
    'customer_id': [random.randint(5000, 5100) for _ in range(200)]
}

df = pd.DataFrame(data)
print(df.head(10))
print(f"\nDataset shape: {df.shape}")
```

Task 1: Basic Grouping and Single Aggregations
Perform the following grouping operations:

    Calculate total sales amount for each region using groupby() and sum()

    Count the number of transactions for each product category using groupby() and count()

    Calculate the average sales amount per salesperson using groupby() and mean()

    For each result, use reset_index() to convert the result into a clean DataFrame
    
    Sort the regional sales results in descending order using sort_values(ascending=False) to
    identify the top-performing region

```python
total_sales_amount=df.groupby('region')['sales_amount'].sum().sort_values(ascending=False).reset_index()
print(total_sales_amount)

top_performing_region = df.groupby('region')['sales_amount'].sum().idxmax()
print(f"The top-performing region is: {top_performing_region}")
```

```python
no_of_trans_per_prod = df.groupby('category')['sales_amount'].count().reset_index()
no_of_trans_per_prod = no_of_trans_per_prod.rename(columns={'sales_amount': 'transaction_count'})
print(no_of_trans_per_prod)
```

```python
avg_sales_per_salesman=df.groupby('salesperson')['sales_amount'].mean().reset_index()
avg_sales_per_salesman=avg_sales_per_salesman.rename(columns={'sales_amount':'avg_sales_amount'})
print(avg_sales_per_salesman)
```

Task 2: Multi-Column Grouping and Multiple Aggregations
Perform advanced grouping analysis:

    1.Group by both region AND category to calculate total sales for each combination using:
      df.groupby(['region', 'category'])['sales_amount'].sum().reset_index()

    2.For each salesperson, calculate three metrics simultaneously using the agg() method:
        Total sales ('sum')
        Average sales ('mean')
        Number of transactions ('count')
    Use this syntax:
    df.groupby('salesperson')['sales_amount'].agg(['sum', 'mean', 'count']).reset_index()
    
    3.Sort the salesperson results by total sales in descending order to identify the top performer
    4.Use .idxmax() on the grouped category sales to find which category has the maximum total revenue

```python
total_sales_combo=df.groupby(['region','category'])['sales_amount'].sum().reset_index()
print(total_sales_combo)
```

```python
salesperson_results=df.groupby('salesperson')['sales_amount'].agg(['sum', 'mean', 'count']).reset_index()

print(salesperson_results)
```

```python
salesperson_results_sorted= salesperson_results.sort_values(ascending=False,by='sum').reset_index()
print(salesperson_results)
print(salesperson_results_sorted)
```

```python
max_total_revenue_category=df.groupby('category')['sales_amount'].sum().idxmax()
print(max_total_revenue_category)
```

Create a comprehensive sales analysis report:

Define a custom aggregation function that calculates the sales range (max - min) for each group:

    def sales_range(x):

Apply this custom function along with standard aggregations to analyze sales by region:

    df.groupby('region')['sales_amount'].agg(['sum', 'mean', 'min', 'max', sales_range]).reset_index()


Create a final summary report that shows for each region:

    Total number of transactions (using customer_id with count)
    Total sales amount
    Average transaction value
    Use dictionary syntax in agg():

    df.groupby('region').agg({
    'sales_amount': ['sum', 'mean'],
    'customer_id': 'count'
    }).reset_index()

Explain the multi-level column structure that results from the dictionary-based aggregation and how it differs from single aggregations



```python
def sales_range(x):
  return x.max() - x.min()

sales_region=df.groupby('region')['sales_amount'].agg(['sum', 'mean', 'min', 'max', sales_range]).reset_index()
print(sales_region)
```

```python
report=df.groupby('region').agg({
    'sales_amount': ['sum', 'mean'],
    'customer_id': 'count'
}).reset_index()

report
```

Okay, let's use a cooking example to understand multi-level columns!

Imagine you're preparing a complex meal with multiple dishes, and you're tracking the ingredients and cooking times:

*   **Single Aggregation (e.g., finding the `total_prep_time` for a specific `recipe`):** This is like looking at a single recipe card and seeing just one total number, like 'Total Prep Time: 30 minutes'. You get one simple piece of information for that item.

*   **Multi-level Columns (like grouping by `meal_course` and then by `dish_type` to find `total_ingredients` and `average_cook_time`):** This is like having a detailed cooking plan:

    You have main sections for `Appetizers`, `Main Courses`, and `Desserts` (these are your primary columns).

    *   Inside the `Main Courses` section, you might have categories like `Pasta Dishes` and `Roasted Meats` (these could be a level of grouping).

    *   Then, inside the `Pasta Dishes` category, you have specific recipes like `Spaghetti Carbonara` and `Lasagna` (another level of grouping).

    *   For each `Spaghetti Carbonara` entry, you want to know two things: the `Total Number of Ingredients` and the `Average Cook Time`. So, in your cooking plan, you'd see:
        *   **Main Courses**
            *   **Pasta Dishes**
                *   **Spaghetti Carbonara**
                    *   Total Ingredients: 7
                    *   Average Cook Time: 15 minutes

    In the DataFrame, this means you'd see `meal_course` as a top-level column, then `dish_type` underneath it, and then `total_ingredients` and `average_cook_time` as the specific metrics for each `dish_type` within each `meal_course`. It's a structured way to look at different characteristics (like ingredients and cook time) for various dishes grouped by their course and type.
